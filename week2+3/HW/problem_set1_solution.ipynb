```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statistics
from collections import Counter
import math
from scipy import stats

data = [13, 15, 16, 16, 19, 20, 20, 21, 22, 22, 22, 25, 25, 25, 25, 30, 30, 33, 33, 35, 35, 35, 35, 35, 36, 40, 45, 46, 52, 70]

# a. Mean and median
mean = statistics.mean(data)
median = statistics.median(data)
print("a.\n")
print(f'mean: {mean}')
print(f'median: {median}')

# b. mode and modality
mode = statistics.mode(data)
X = Counter(data).items()
print("b.\n")
for value, count in X:
    print(f'{value}: {count}')
print(f'mode: {mode}')
max_count = max(Counter(data).values())
modal_value = [value for value, count in X if count == max_count]
if len(modal_value) == 1:
    print("unimodal")
elif len(modal_value) == 2:
    print("bimodal")
elif len(modal_value) == 3:
    print("trimodal")
else:
    print('multimodal')

# c. Midrange
midrange = (max(data) + min(data))/2
print('c. \n')
print(f'Midrange: {midrange}')

# d. Q1 and Q3 
Q_1 = np.percentile(data, 25)
Q_3 = np.percentile(data, 75)
print('d\n')
print(f"Q1: {Q_1}")
print(f'Q3: {Q_3}')

# e. Five-number summary
print('e\n')
print(f"Min: {min(data)}, Q1: {Q_1}, Median: {median}, Q3: {Q_3}, Max: {max(data)}")

# f. Boxplot
print('f.\n')
plt.boxplot(x=data, meanline='red')
plt.xlabel('Age')
plt.title('Boxplot of Age')
IQR = Q_3 - Q_1
print(Q_1 - 1.5*IQR)
print(Q_3 + 1.5*IQR)

# g. Variance and standard deviation
print('g.\n')
sum_data = 0
for i in data:
    sum_data = sum_data + i**2
variance = ((1/len(data)) * sum_data) - mean*mean
standard_deviation = math.sqrt(variance)
print(f'variance: {variance}') # or np.var(data, ddof=1)
print(f'standard deviation: {standard_deviation}')
```

```python
age = [23, 23, 27, 27, 39, 41, 47, 49, 50, 52, 54, 54, 56, 57, 58, 58, 60, 61]
percent_fat = [9.5, 26.5, 7.8, 17.8, 31.4, 25.9, 27.4, 27.2, 31.2, 34.6, 42.5, 28.8, 33.4, 30.2, 34.1, 32.9, 41.2, 35.7]

# a. mean, median, variance and standard deviation
print('a.\n')
# age
mean_age = statistics.mean(age)
median_age = statistics.median(age)
variance_age = np.var(age)
standard_deviation_age = np.std(age)
print(f'mean of age: {mean_age}')
print(f'median of age: {median_age}')
print(f'variance of age: {variance_age}')
print(f'standard deviation of age: {standard_deviation_age}')

# percent fat
mean_fat = statistics.mean(percent_fat)
median_fat = statistics.median(percent_fat)
variance_fat = np.var(percent_fat)
standard_deviation_fat = np.std(percent_fat)
print(f'mean of $fat: {mean_fat}')
print(f'median of %fat: {median_fat}')
print(f'variance of %%fat: {variance_fat}')
print(f'standard deviation of %fat: {standard_deviation_fat}')


# b. Boxplot
box = plt.boxplot([age, percent_fat],
                  labels=['Age', '% Fat'],
                  boxprops=dict(color='blue'),  # Màu viền box
                  medianprops=dict(color='red'),  # Màu đường median
                  whiskerprops=dict(color='green'),  # Màu whiskers
                  capprops=dict(color='orange'))  # Màu caps
plt.title('boxplots for age and %fat')
plt.ylabel('Values')
```

```python
# c. scatter plot
plt.scatter(age, percent_fat)
plt.xlabel("Age")
plt.ylabel("%fat")
plt.title("scatter plot for these two variables")
```

```python
# d. Calculate the Pearson and Spearman correlation coefficients between the two  attributes.
Pearson_correlation_coefficients = stats.pearsonr(age, percent_fat)
Spearman_correlation_coefficients = stats.spearmanr(age, percent_fat)
print(f'Pearson correlation coefficients: {Pearson_correlation_coefficients}')
print(f'Spearman correlation coefficients: {Spearman_correlation_coefficients}')
```
